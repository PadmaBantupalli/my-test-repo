```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, fcluster
import numpy as np

# Data: Time spent on social media and gym
data = np.array([[2, 3],  # Alan
                 [5, 2],  # Lisa
                 [5, 3],  # Joe
                 [1, 4],  # Max
                 [4, 5]])  # Caro

labels = ['Alan', 'Lisa', 'Joe', 'Max', 'Caro']

# Perform hierarchical clustering
linked = linkage(data, method='ward')

# Assign cluster labels using a distance threshold
cluster_labels = fcluster(linked, t=3, criterion='distance')
print(cluster_labels)

# Scatter plot with labels annotated
plt.figure(figsize=(8, 5))
for i, point in enumerate(data):
    plt.scatter(point[0], point[1], label=labels[i])
    plt.annotate(f"{labels[i]} (Cluster {cluster_labels[i]})", (point[0], point[1]), textcoords="offset points", xytext=(10, 10), ha='center')

plt.title("Scatter Plot with Clusters")
plt.xlabel("Social Media Time")
plt.ylabel("Gym Time")
plt.legend()
plt.show()
```
