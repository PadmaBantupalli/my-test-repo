```python
import pandas as pd

df = pd.read_csv("./Data/titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
print("************")
print(df.isnull().any())
print("************")
print(df['Embarked'].unique())
print("************")
print(df.columns[df.isna().any()])
```

```python
values = {'Age': df.Age.mean(), 'Embarked': df.Embarked.mode()[0] }
#df['Age'].fillna(df['Age'].mean(), inplace=True)
input_df = df.fillna(value=values)
print(input_df.Age.isnull().any())
print(input_df.Embarked.isnull().any())
```

```python
input_df.Embarked.isnull().sum()
```

```python
input_df = pd.get_dummies(input_df, columns=['Sex', 'Embarked'], dtype=int)
#pd.get_dummies(input_df.Sex)
```

```python
input_df.head()
```

```python
y = input_df['Survived']
input_df.drop(['Sex_male', 'Embarked_S', 'Survived'], axis=1, inplace=True)
```

```python
y.head()
```

```python
df.head()
```

```python
df.columns[df.isna().any()]
```

```python
df.dropna(subset=['Age','Embarked'], inplace=True)
df.columns[df.isna().any()]
```

```python
input_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(input_df, y, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB

Gnb = GaussianNB()
```

```python
Gnb.fit(X_train, y_train)
```

```python
Gnb.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

cv = cross_val_score(GaussianNB(), input_df, y, cv=5)
cv
```
