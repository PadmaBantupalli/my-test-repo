```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
# digits.data is a numpy array
digits.data
```

```python
#create a data frame for the data
inputs = pd.DataFrame(digits.data)
inputs.head()
```

```python
#Add the target to the inputs dataframe
inputs['target'] = digits.target
inputs.head()
```

```python
target = inputs['target']
target.head()
```

```python
inputs.isnull().any()
inputs.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray()
for i in range(3):
    plt.imshow(digits.images[i])
    plt.show()
```

```python
inputs.drop('target', axis='columns', inplace=True)
```

inputs.head()

```python
inputs.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=20)
```

```python
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
